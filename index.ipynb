## The Question

Which advertising channel has a strong relationship with sales volume, and can be used to model and predict the sales. 

### Step 1: Read the dataset and inspect its columns and 5-point statistics

```python
# Load necessary libraries and import the data
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
# Check the columns and first few rows
df = pd.read_csv('Advertising.csv',index_col=0)

df.describe()
# df.head()
```

```python
# there are 200 rows, each of which represent an advertising campaign
# in each row is the $ spent on each channel and the total # of total sales


```

### Step 2: Plot histograms with kde overlay to check for the normality of the predictors

```python
# For all the variables, check if they hold normality assumption

for i in df: 
    df[i].plot.hist(density=True, label = i.title()+' histogram')
    df[i].plot.kde(label =i.title()+' kde')
    plt.legend()
    plt.show()
```

```python
# Record your observations on normality here 

# sales seems the most normal
# newspaper seems the least normal
# radio and tv are normal-ish so still can be used as a predictors
# all are positively skewed (some more than others), except tv, which is negatively skewed

```

### Step 3: Test for the linearity assumption. 
Use scatterplots to plot each predictor against the target variable

```python
# visualize the relationship between the preditors and the target using scatterplots
# target variables = on the y-axis
# target var here = sales


fig, axs = plt.subplots(1, 3, sharey=True, figsize=(18, 6))
for idx, channel in enumerate(['TV', 'radio', 'newspaper']):
    df.plot(kind='scatter', x=channel, y='sales', ax=axs[idx], label=channel)
plt.legend()
plt.show()

```

```python
# looks like newspaper has the least-linear relationship with sales, while tv has the most
# newspaper also seems to have the highest variance between its datapoints in relation to sales, followed by radio and then tv
# newspaper doesn't seem like a good parameter to use in a linear regression analysis because it is not really linear
```

### Conclusion so far !

Based on above initial checks, we can confidently say that TV and radio appear to be good predictors for our regression analysis. Newspaper is very heavily skewed and also doesnt show any clear linear relationship with the target. 
> We shall move ahead with our analysis using TV and radio , and count out the newspaper due to the fact that data violates ols assumptions

Note: Kurtosis can be dealt with using techniques like log normalization to "push" the peak towards the center of distribution. We shall talk about this in the next section. 

### Step 4: Run a simple regression in `statsmodels` with TV as a predictor

```python
# import libraries


# build the formula 
# y~x

f = 'sales~TV'

# create a fitted model in one line
model = ols(formula=f, data=df).fit()
```

### Step 5: Get regression diagnostics summary


```python
model.summary()
```

### Record your observations on "Goodness of fit"


Note here that the coefficients represent associations, not causations


```python
# since r^2 is 0.612, I would say that our regression line is a good fit because this # is close to 1
# since our p-value is so low (0.0), i would also say that we can reject the null hypothesis in good faith and conclude that putting $ towards TV marketing is related to sales
# for unit increase in tv spending, sales units increase 0.0475x
```

### Step 6:  Draw a prediction line with data points omn a scatter plot for X (TV) and Y (Sales)

Hint: We can use `model.predict()` functions to predict the start and end point of of regression line for the minimum and maximum values in the 'TV' variable. 

```python
# create a DataFrame with the minimum and maximum values of TV
Xnew = pd.DataFrame({'TV': [df.TV.min(),df.TV.max()]})


# make predictions for those x values and store them
Ynew = model.predict(Xnew)


# first, plot the observed data and the least squares line
df.plot(kind='scatter', x='TV', y='sales') # plotting regular data from original dataframe
plt.plot(Xnew, Ynew, c='red', linewidth=1) # plotting new x and y values with matplotlib with a OLS line that goes through the new x,y point we are predicting the position of

```

### Step 7: Visualize the error term for variance and heteroscedasticity

```python
# import scipy.stats as stats
# residuals = model.resid
# fig = sm.graphics.qqplot(residuals, dist=stats.norm, line='45', fit=True)
# fig.show()



fig = plt.figure(figsize=(18,9))
fig = sm.graphics.plot_regress_exog(model, "TV", fig=fig)
plt.show()

```

```python
# Record Your observations on residuals

# the Residuals versus TV graph shows the residuals to be heteroscedastic
# heteroscedasity indicates that an assumption for linear regression has been violated



# the qq plot (Y and Fitted vs. X) shows that the data is normally distributed
# the partial regression plot shows the relationship to be linear
# the regression plot w/the other independent vars taken into account also shows the data to be linear
```

### Next, repeat above for radio and go through the same process, recording your observations

```python
f = 'sales~radio'

# create a fitted model in one line
model = ols(formula=f, data=df).fit()


# create a DataFrame with the minimum and maximum values of TV
Xnew = pd.DataFrame({'radio': [df.radio.min(),df.radio.max()]})


# make predictions for those x values and store them
Ynew = model.predict(Xnew)


# first, plot the observed data and the least squares line
df.plot(kind='scatter', x='radio', y='sales') # plotting regular data from original dataframe
plt.plot(Xnew, Ynew, c='red', linewidth=1) # plotting new x and y values with matplotlib with a OLS line that goes through the new x,y point we are predicting the position of


fig = plt.figure(figsize=(18,9))
fig = sm.graphics.plot_regress_exog(model, "radio", fig=fig)
plt.show()

model.summary()
```

```python
# Record your observations here for goodnes of fit 

# seems that everything is pretty much the same as it was for tv
# except that radio has a smaller r^2 value, indicating that our model
# fits this data less well than it fits the 'sales~TV' relationship
# it is leptokurtic (kurtosis > 3)
# it is skewed negatively indicating that the 
# mean and median will be less than the mode
# A "unit" increase in radio spending is associated with a 0.2025 "unit" increase in Sales.
```
